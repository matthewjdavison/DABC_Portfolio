```python
def display_column_data(value):
    print(f'{value}\n')
    print('\nMode')
    display(ad_tv_df[value].mode()[0])
    print('\nHead (3)')
    display(ad_tv_df[value].head(3))
    print('\nInfo')
    display(ad_tv_df[value].info())
    return

def display_multi_column_data(value):
    for column in value:
        display_column_data(column)
    return
```

Those 4 columns seem to contain the text 'UNKNOWN' when there is missing data.
I will convert those to nan values, so they're floats. 

```python
ad_tv_df[ad_tv_df['Score'] == 'UNKNOWN']
```

```python
def convert_unknown_to_nan(value):
    if value == 'UNKNOWN':
        value = np.nan
    return value
```

```python
ad_tv_df['Score'] = ad_tv_df['Score'].apply(convert_unknown_to_nan)
ad_tv_df['Score'] = pd.to_numeric(ad_tv_df['Score'])
```

```python
ad_tv_df['Score'].info()
```

```python
ad_tv_df['Episodes'] = ad_tv_df['Episodes'].apply(convert_unknown_to_nan)
ad_tv_df['Episodes'] = pd.to_numeric(ad_tv_df['Episodes'])
```

```python
ad_tv_df['Episodes'].info()
```

```python
ad_tv_df['Rank'] = ad_tv_df['Rank'].apply(convert_unknown_to_nan)
ad_tv_df['Rank'] = pd.to_numeric(ad_tv_df['Rank'])
```

```python
ad_tv_df['Rank'].info()
```

```python
ad_tv_df['Scored By'] = ad_tv_df['Scored By'].apply(convert_unknown_to_nan)
ad_tv_df['Scored By'] = pd.to_numeric(ad_tv_df['Scored By'])
```
